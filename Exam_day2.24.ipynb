```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(-4.5,5,0.5)
y=2*x**2
plt.plot(x,y)
plt.show()
```

```python
import platform
from matplotlib import font_manager, rc # Runtime Configuration

# 맑은고딕 선택
path = "c:/Windows/Fonts/malgun.ttf"
# Darwin은 맥북을 의미.
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')

elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')
```

```python
x=np.arange(-3,4,1)
y1=x
y2=x+1
y3=x+2
y4=x+3
plt.plot(x,y1,'>--r',x,y2,'s-g',x,y3,'d:b',x,y4,'-.Xc')
plt.legend(['자료1','자료2','자료3','자료4'],loc=2)
plt.xlabel('x축')
plt.ylabel('Y축')
plt.yticks(fontname='Arial')
plt.xticks(fontname='Arial')
plt.grid()
plt.title('그래프 제목')
plt.show()
```

```python
height=[165,177,160,180,185,155,172]
weight=[62,67,55,74,90,43,64]
print(np.arange(1,8))
size=100*np.arange(1,8)
colors=['r','g','b','c','m','k','y']
plt.scatter(height,weight,s=size,c=colors)
plt.xlabel('Height(m)')
plt.ylabel('Weight(kg)')
plt.title('Height(m) & Weight(kg)')
plt.grid()
plt.show()
```
